```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
total_revenue= df.groupby("Product")["Revenue"].sum()
print(total_revenue)
```

```python
avg_units= df.groupby("Product")["Units_Sold"].sum()
print(avg_units)
```

```python
#line chart

pivot_trend= df.pivot(index="Month",columns="Product", values="Revenue")

pivot_trend.plot(kind="line", marker="o", linewidth=2)
plt.title("Product vs Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation= 45)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(title="Product")
plt.tight_layout()
plt.show()
```

```python
#bar chart

total_rev= df.groupby("Product")["Revenue"].mean()
total_rev.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Total Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation= 45)
plt.show()
```

```python
# pie chart

prod_revenue= df.groupby('Product')["Revenue"].sum()/100
plt.figure(figsize=(4,4))
plt.pie(prod_revenue, labels=prod_revenue.index,autopct='%1.1s%%', startangle=140)
plt.title('Revenue Contribution of each Product')
plt.show()
```
